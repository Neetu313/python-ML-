```python
#
from sklearn.datasets import make_classification, make_regression
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import plotly.graph_objs as go
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_features = 10, 
                           n_classes = 2)

scalar = StandardScaler()         # scale the data
scalar.fit(X)
X_ = scalar.transform(X)

pca = PCA(n_components = 2)
pca.fit(X_)
X__ = pca.transform(X_)
```

```python
fig = go.Figure(data = [go.Scatter(x = X_[:, 0], 
                                   y = X_[:, 1],
                                   mode =  'markers',
                                   marker = {'size' : 6,      
                                  'color' : y})])
fig.show()
```

```python
# with whiting


scalar = StandardScaler()         # scale the data
scalar.fit(X)
X_ = scalar.transform(X)
pca = PCA(n_components = 2, whiten = False)
pca.fit(X_)
X__ = pca.transform(X_)
fig = go.Figure(data = [go.Scatter(x = X_[:, 0], 
                                     y = X_[:, 1],
                                    
                                     mode =  'markers',
                                     marker = {'size' : 6,      
                                     'color' : y})])
fig.show()
```

```python
pca = PCA(n_components = 2, whiten = True)
pca.fit(X_)
X__ = pca.transform(X_)
fig = go.Figure(data = [go.Scatter(x = X_[:, 0], 
                                     y = X_[:, 1],
                                    
                                     mode =  'markers',
                                     marker = {'size' : 6,      
                                     'color' : y})])
fig.show()
```

```python
# create own data set  and apply whitening
import numpy as np
X = np.linspace(-10, 10, 1000) + 5*np.random.random(1000)
Y = np.linspace(-10, 10, 1000) + 5*np.random.random(1000)
```

```python
go.Figure(data = [go.Scatter(x = X,
                             y = Y,
                            mode = 'markers')])
```

```python
X = np.asarray((X, Y)).T.shape
```

```python
scalar = StandardScaler()        
scalar.fit(X)
X_ = scalar.transform(X.reshape(-1, 1))
pca = PCA(n_components = 1, whiten = False)
pca.fit(X_)
X__ = pca.transform(X_)
go.Figure(data = [go.Scatter(x = X__[:, 0])])
```
