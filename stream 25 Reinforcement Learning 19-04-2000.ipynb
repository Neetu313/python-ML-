```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from plotly.subplots import make_subplots
import plotly.graph_objs as go
```

```python
X, y = make_moons(n_samples = 1000,
                  noise = 0.05)
```

```python
fig = go.Figure()
fig.add_scatter(x = X[:, 0],
                y = X[:, 1],
                mode = 'markers',
                marker_color = y)
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
y_pre_kmeans = kmeans.labels_
```

```python
agg = AgglomerativeClustering(n_clusters = 2)
agg.fit(X)
y_pre_agg = agg.labels_
```

```python
dbs = DBSCAN(eps = 0.1, min_samples = 1)
dbs.fit(X)
y_pre_dbs = dbs.labels_
```

```python
fig = make_subplots(rows = 1, cols = 3)
fig.add_scatter(x = X[:, 0],
                y = X[:, 1],
                mode = 'markers',
                marker_color = y_pre_kmeans, row = 1,
                col = 1)
fig.add_scatter(x = X[:, 0],
                y = X[:, 1],
                mode = 'markers',
                marker_color = y_pre_agg,
                row = 1, col = 2)
fig.add_scatter(x = X[:, 0],
                y = X[:, 1],
                mode = 'markers',
                marker_color = y_pre_dbs,
                row = 1, col = 3)
```
